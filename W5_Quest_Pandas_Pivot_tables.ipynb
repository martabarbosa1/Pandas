# PIVOT TABLE

## It's like in Excel, but with Pandas

There are 2 syntaxes, choose your preferred :
- As a method : YOUR_DATAFRAME.pivot_table()
- As a function : pd.crosstab()

```
# Example
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', aggfunc = 'mean')
```

## Create a Pivot table with :
- Months on rows
- the maximum of Max Temperature on values
- Opinion on columns

```
# Hop !
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', columns='OPINION', aggfunc='max')

#On month 1: for the bad opinion the maximum temperature was 11, for the very bad opinion the maximum temperature was 10
```

## Pivot table with :
- Days on rows
- Precipitation on values
- Months on columns


```
# Hop !
df_weather.pivot_table(values='PRECIP_TOTAL_DAY_MM', index='DAY', columns='MONTH')
```

## Pivot table with :
- Months on rows
- the sum of Precipitations on values

**Then display it as a column chart.** (optionnal)

```
# Hop
df_pivot=df_weather.pivot_table(values='PRECIP_TOTAL_DAY_MM', 
                                index='MONTH', 
                                aggfunc='sum')
df_pivot
```

```
#using matplotlib
ax=df_pivot.plot(xticks=df_pivot.index, ylabel='PRECIP_TOTAL_DAY_MM')
```

```
#using seaborn
sns.barplot(x=df_pivot.index, y="PRECIP_TOTAL_DAY_MM", data=df_pivot)
```

## Pivot table with :
- Months on rows
- the average of Max Temperature on a first column of values
- the average of Sunhours on a second column of values

```
# Hop !
df_pivot=df_weather.pivot_table(values=['MAX_TEMPERATURE_C', 'SUNHOUR'], 
                       index='MONTH',
                       aggfunc='mean')
df_pivot
```

# Chart
Using your favorite library (MatPlotLib, Seaborn, Plotly or what you want), draw a chart from the previous pivot table with :
- Temperature in a blue column chart

Tip : Once your pivot table created, you can help yourself with the `.reset_index()`  method.

```
# Your code here
df_pivot1=df_pivot.reset_index()
df_pivot1
```

```
ax=sns.barplot(data=df_pivot1, x='MONTH', y='MAX_TEMPERATURE_C', color='lightblue')
ax2=ax.twinx()
sns.lineplot(data=df_pivot1, x='MONTH', y='SUNHOUR', ax=ax2, color='red')
```
